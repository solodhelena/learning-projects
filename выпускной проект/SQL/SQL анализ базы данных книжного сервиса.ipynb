## Задачи.

### Посчитайте, сколько книг вышло после 1 января 2000 года;

```python
pd.io.sql.read_sql("SELECT * FROM books LIMIT 10;", con = engine) 
```

```python
pd.io.sql.read_sql(
    "SELECT COUNT(book_id) FROM books WHERE publication_date > '2000-01-01';", con = engine) 
```

После 1 января 2000 года вышло 819 книг.

### Для каждой книги посчитайте количество обзоров и среднюю оценку.

```python
pd.io.sql.read_sql(
    '''SELECT b.title, 
    COUNT(DISTINCT r.review_id), 
    AVG(rat.rating) 
    FROM reviews AS r 
    LEFT JOIN books AS b ON b.book_id = r.book_id 
    LEFT JOIN ratings AS rat ON b.book_id = rat.book_id 
    GROUP BY b.title
    ORDER BY COUNT(DISTINCT r.review_id) DESC;''', con = engine)
```

```python
pd.io.sql.read_sql('''
SELECT 
    b.title AS title,
    b.book_id AS id,
    AVG(r.rating) AS avg_rating,
    COUNT(DISTINCT rev.text)
FROM
    books AS b
INNER JOIN ratings AS r ON r.book_id = b.book_id
INNER JOIN reviews AS rev ON rev.book_id = b.book_id
GROUP BY
     b.title,
     b.book_id
ORDER BY
      COUNT(DISTINCT rev.text) DESC
limit 20
''', con = engine)
```

Есть ряд книг с единичными обзорами. Наибольшее количество обзоров - 7. Средняя оценка 3,66.

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры.

```python
pd.io.sql.read_sql('''SELECT p.publisher,
COUNT(b.title) 
FROM books AS b 
LEFT JOIN publishers AS p ON b.publisher_id = p.publisher_id 
WHERE num_pages > 50 
GROUP BY p.publisher 
ORDER BY COUNT(b.title) DESC
LIMIT 15;''', con = engine)
```

Издательство, которое выпустило наибольшее число книг более 50 страниц - Penguin Books, на втором месте Vintage.

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками.

```python
pd.io.sql.read_sql('''SELECT
      data.author AS author,
      AVG(data.avg_rating) AS avg_rating
FROM
(SELECT
     a.author AS author,
     b.book_id AS id,
     AVG(r.rating) AS avg_rating,
     COUNT(r.rating) AS count_rating
     FROM authors AS a
     INNER JOIN books AS b ON a.author_id = b.author_id
     INNER JOIN ratings AS r ON r.book_id = b.book_id
     GROUP BY author, id
     HAVING COUNT(r.rating) >= 50) AS data
GROUP BY
    author
ORDER BY
    avg_rating DESC;''', con = engine)
```

Автор с самой высокой средней оценкой книг - J.K. Rowling. Не удивительно, Гарри Поттер уже долгое время на слуху, а в те годы был пик его популряности. 

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
pd.io.sql.read_sql('''
SELECT ROUND(avg(count)) 
FROM (SELECT COUNT(text)
FROM reviews
FULL OUTER JOIN (SELECT username, COUNT(rating_id) as cnt
FROM ratings
GROUP BY username) AS cnt_rating ON cnt_rating.username = reviews.username
WHERE cnt > 50
GROUP BY reviews.username) as count;''', con = engine)                                
```

Среднее количество обзоров от авторов поставивших более 50 оценок - 24.
