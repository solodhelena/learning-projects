### Исследуем соотношение видов объектов общественного питания по количеству. Построим график.

```python
# Сгруппируем таблицу для построения графика по количеству объектов. Посчитаем количество каждого типа объекта. 

count_id = rest_data.groupby('object_type')["id"].count()
count_id = pd.DataFrame({'object_type':count_id.index, 'count':count_id.values})
count_id = count_id.sort_values(by='count', ascending=False)
count_id
```

```python
# Посторим график распределения количества объектов по типам объектов.

fig, ax = plt.subplots(figsize=(15, 5))

sns.set_style('whitegrid')

sns.barplot(data=count_id, x='object_type', y='count', linewidth = 0.7, alpha = 1).set_title("Распределение количества объектов общественного питания по типам объектов", fontsize=16)

ax.set(xlabel='Тип объекта', ylabel='Количество объектов')

for container, number in zip(ax.containers, count_id['count']):
    ax.bar_label(container, labels=count_id['count'])

plt.show()
```

Итак, на первом месте кафе - почти 6 тысяч объектов, далее столовая - 2409, ресторан - 2208, предприятия быстрого питания - FastFood - 1829, следующий объект - бар - уже в два раза меньше по количеству - 826 объектов.

Распределение в процентах: на первом месте 39.7% - кафе, далее столовая - 16,5%, затем рестораны - 15%, FastFood - 12.7%.

### Для какого вида объекта общественного питания характерно сетевое распространение?

```python
# Построим сводную таблицу по типам объектов и их типу: сетевые, не сетевые, посчитаем их количество.

pivot_chain = rest_data.pivot_table(index=['object_type', 'chain'], values='id', aggfunc='count')

pivot_chain
```

Сетевую модель лучше открывать по типу кафе, предприятия быстрого обслуживания или ресторана. Также необходимо учитывать, что количество посадочных мест не должно превышать 50 для предприятий быстрого питания и кафе, и не должно быть больше чем на 70-100 мест для ресторанов. При сетевой модели лучше открывать больше объектов с небольшим количеством мест, чем мало точек с большим количеством посадочных мест. Сетевая модель направлена на покрытие больших районов города, нежели единичные объекты.
